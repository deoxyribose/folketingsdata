```python
from pydub import AudioSegment
import pandas as pd
import json
import os
import csv
```

```python
path = '/home/folzd/Desktop/'
#path = '/home/folzd/Tale/'
```

```python
%%bash
for file in *.mp3; do ffmpeg -i "$file" -acodec pcm_s16le -ac 1 -ar 16000 "${file/mp3/wav}";done
```

```python
audiofiles = [f for f in os.listdir(path) if f.endswith('.wav')]
jsonfiles = [f for f in os.listdir(path) if f.endswith('.json')]
```

```python
zip(sorted(audiofiles),sorted(jsonfiles))
```

```python
def writeCSV(audiofile,jsonfile):
    book = AudioSegment.from_mp3(audiofile)

    with open(jsonfile) as f: 
        syncmap = json.loads(f.read())

    sentences = []
    for fragment in syncmap['fragments']:
        if ((float(fragment['end'])*1000) - float(fragment['begin'])*1000) > 500:
            sentences.append({"audio":book[float(fragment['begin'])*1000:float(fragment['end'])*1000], "text":fragment['lines'][0]})

    # save to pandas 
    df = pd.DataFrame(columns=['wav_filename','wav_filesize','transcript'])

    #create directory
    name = audiofile[:-4]
    os.makedirs(name)

    # export audio segment
    for idx, sentence in enumerate(sentences):
        text = sentence['text'].lower()
        print(text)
        sentence['audio'].export(name+"/sample-"+str(idx)+".wav", format="wav")
        temp_df = pd.DataFrame([{'wav_filename':"sample-"+str(idx)+".wav",'wav_filesize':os.stat(audiofile).st_size,'transcript':text}], columns=['wav_filename','wav_filesize','transcript'])
        df = df.append(temp_df)

    df.to_csv(name+"/sample.csv",index=False,encoding="utf-8",quoting=csv.QUOTE_NONE)
```

```python
for (a,j) in zip(sorted(audiofiles),sorted(jsonfiles)):
    writeCSV(a,j)
```
